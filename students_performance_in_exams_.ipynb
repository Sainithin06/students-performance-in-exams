# Bivariate Visualisation

Bivariate analysis is a kind of statistical analysis when two variables are observed against each other. One of the variables will be dependent and the other is independent. The variables are denoted by X and Y. The changes are analyzed between the two variables to understand to what extent the change has occured.

Bivariate analysis is not the same as two sample data analysis. With two sample data analysis (like a two sample z test in Excel), the X and Y are not directly related. You can also have a different number of data values in each sample; with bivariate analysis, there is a Y value for each X.

# How many students passed Mathematics Exam? (above 60)

We will calculate each student will pass in each exam by make a passmark as an indicator of success and categorize each student within all exam by GPA

we will set the minimum marks to 60 to pass in a exam

```python
passmark = 60

data_port['Pass_Math_Exam'] = (data_port['Math_Score'] > passmark).map({True: "P",False: "F"})

data_port['Pass_Math_Exam'].value_counts()
```

```python
data_port['Pass_Math_Exam'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Statistics of Mathematics scores')

plt.ylabel('')
```

From the chart above, we can see that 64.7% of the student passed mathematics(i.e above 60) while 35.3% failed.

```python
data_port['Pass_Reading_Exam'] = (data_port['Reading_Score'] > passmark).map({True: "P", False: "F"})
                                                                                        
data_port['Pass_Reading_Exam'].value_counts()
```

```python
data_port['Pass_Reading_Exam'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Statistics of Reading scores')

plt.ylabel('')
```

```python
data_port['Pass_Writing_Exam'] = (data_port['Writing_Score'] > passmark).map({True: "P",
                                                     False: "F"})
data_port['Pass_Writing_Exam'].value_counts()
```

```python
data_port['Pass_Writing_Exam'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Statistics of Writing scores')

plt.ylabel('')
```

From the chart above, we can see that 68.5% of the student passed Reading exam(i.e above 60) while 31.5% failed

# How many students pass in all exams?

```python
data_port['Overall_Passed'] = data_port.apply(lambda df:
                                    'Passed_all' if df['Pass_Math_Exam'] == 'P' 
                                    and df['Pass_Reading_Exam']  == 'P' 
                                    and df['Pass_Writing_Exam']  == 'P' 
                                    else 'Less_than_3_passes',
                                    axis=1)
data_port['Overall_Passed'].value_counts()
```

```python
sns.countplot(data= data_port, x= 'Overall_Passed');
```

From the chart above, we can say that about 85% of the student passed the three exams.

# GPA of Students

```python
data_port['Total_Marks'] = data_port['Math_Score'] + data_port['Reading_Score'] + data_port['Writing_Score']
data_port['Percentage'] = data_port['Total_Marks'] / 3.0
data_port.head()
```

```python
def GetGrade(Percentage):
    if Percentage >= 80:
        return 'A'
    elif Percentage >= 70:
        return 'B'
    elif Percentage >= 60:
        return 'C'
    elif Percentage >= 50:
        return 'D'
    else:
        return 'F'
```

```python
data_port['GPA'] = data_port.apply(lambda df : GetGrade(df['Percentage']), axis= 1)
data_port.head()
```

```python
Exams_Performance = data_port['GPA'].value_counts().to_frame()

Exams_Performance
```

```python
GPA_Order = list("ABCDF")
sns.catplot(data= data_port, x= 'GPA',order= GPA_Order, kind= 'count');
plt.title('GPA Analysis Of Students');
```

From the chart, we see that majority of the class had an average GPA of C and B
